Loading Data and importing pandas library

```python
import pandas as pd
df=pd.read_csv("matches.csv")
df.head()
```

Checking Data

```python
df.info()
```

```python
df.isnull().sum()
```

Importing matplotlib library and Analyzing no of match won by each team

```python
import matplotlib.pyplot as plt
df['winner'].value_counts().plot(kind='bar')
plt.title('Number of Matches Won by Each Team')
plt.xlabel('Team')
plt.ylabel('Wins')
plt.xticks(rotation=45)
plt.show()
```

Insight:
Mumbai Indians and Chennai Super Kings have the highest number of wins

Toss Impact

```python
toss_win_match_win = df[df['toss_winner'] == df['winner']]
toss_win_match_win.shape[0]/df.shape[0]*100
```

Insight:
Teams that win the toss also win the match around 50–60% of the time.

Matches per season

```python
df['season'].value_counts().sort_index().plot()
plt.title('Number of Matches per season')
plt.xlabel('Season')
plt.ylabel('Matches')
plt.xticks(rotation=45)
plt.show()
```

Insight:
Number of matches rose steadily till around 2013 then dropped and stabilized later

Top 10 venues

```python
df['venue'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Venues for Matches')
plt.xlabel('Venue')
plt.ylabel('Matches')
plt.xticks(rotation=45)
plt.show()
```

Insight:Eden Garden has hosted the highest no of matches

```python
import seaborn as sns
sns.countplot(x='toss_decision',data=df)
plt.title("Toss decission distribution")
plt.xticks(rotation=45)
plt.show()
```

Insight:Teams prefer fielding over batting after winning the toss.

Winners across season

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.countplot(x='season', hue='winner', data= df)

plt.title("Winners across seasons")
plt.xlabel("Season")
plt.ylabel("Number of Matches Won")
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

No single team dominates; wins are fairly evenly distributed across seasons
